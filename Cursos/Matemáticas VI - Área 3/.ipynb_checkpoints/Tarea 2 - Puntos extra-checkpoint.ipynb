```python
from pandas import DataFrame
from math import log
from fractions import Fraction
```

```python
def armonica(n):
    assert type(n)==type(1) and n>=1
    serie=[Fraction(1,k) for k in range(1,n+1)]
    return sum(serie)
```

```python
tablita=DataFrame({'Serie armónica':[armonica(i) for i in range(1,1001)],
                   'Logaritmo':[log(i) for i in range(1,1001)]},
                    index=[i for i in range(1,1001)])

print(tablita.head(10))
```

```python
tablita.astype(float).plot(figsize=(12,5),use_index=True,grid=True)
```

Ambas gráficas se parecen, pero tienen un _desfase_. Si se avanzara más a la derecha, la distancia (vertical) entre ambas curvas se aproximaría a la constante de [Euler-Mascheroni](https://en.wikipedia.org/wiki/Euler%E2%80%93Mascheroni_constant):
\begin{equation}
\gamma=\lim_{n\to\infty}\left(S_n-s_n\right)=
    \lim_{n\to\infty}\left(\sum_{k=1}^n \frac{1}{k}-\ln(n)\right)\approx 0.5772156649
\end{equation}

```python
print('La diferencia entre las curvas en el milésimo término es {:.10f}'.format(
    -tablita.iloc[-1].astype(float).diff().values[-1]))
```
