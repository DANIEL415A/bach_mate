* ¿Cómo generar la lista de la suma acumulada de los términos de la progresión armónica $a_n=\{\frac{1}{n}\}_{n=1}^{\infty}$?
\begin{equation}
    S_1=a_1 \\
    S_2=a_1+a_2 \\
    S_3=a_1+a_2+a_3 \\
    \dots\\
    S_n=\sum_{j=1}^n a_j=1+\frac{1}{2}+\cdots+\frac{1}{n}\\
\end{equation}

```python
# Función que me regresa los términos de la armónica.
def armonica(n):
    assert type(n)==type(0) and n >=1
    return Fraction(1,n)

print(armonica(3))
```

```python
#Hacer la lista de las diez primeras sumas
limite=1000
lista_chula=[(sum([armonica(i) for i in range(1,j)])) for j in range(1,limite)]
print(lista_chula)
```

```python
[float(l) for l in lista_chula]
```

## Graficar listas de números

Toman un plano cartesiano; el eje horizontal (el de las $X$) contendrá los índices de la sucesión: $0, 1, 2, \dots$. El eje vertical contendrá los elementos de la sucesión.

```python
import matplotlib.pyplot as plt
```

```python
aritmetica=[2*i-5 for i in range(30)] #a(0),a(1),...,a(29) (30 numeros en total)
plt.plot(aritmetica,'.')
plt.grid()
```
